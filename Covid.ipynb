```python
import numpy                             
import matplotlib.pyplot as plt
import pandas as pd
```

```python
link_covid_data= 'https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv'    #buat variabel untuk memasukan link data
raw_data = pd.read_csv(link_covid_data)                   
raw_data.head()
```

```python
indonesia = raw_data[raw_data['location'] == 'Indonesia']
indonesia
```

```python
turkey = raw_data[raw_data['location'] == 'Turkey']
turkey
```

```python
poland = raw_data[raw_data['location'] == 'Poland']
poland
```

```python
argentina = raw_data[raw_data['location'] == 'Argentina']
argentina
```

```python
indonesia2 = indonesia[indonesia['date'] >= '2020-02-24']         #variabel untuk menentukan tanggal
turkey2 = turkey[turkey['date'] >= '2020-02-24']
poland2 = poland[poland['date'] >= '2020-02-24']
argentina2 = argentina[argentina['date'] >= '2020-02-24']
```

```python
plt.plot(indonesia2['date'], indonesia2['total_deaths'])          #plot negara untuk menentukan jujmlah kematian
plt.plot(turkey2['date'], turkey2['total_deaths'])
plt.plot(poland2['date'], poland2['total_deaths'])
plt.plot(argentina2['date'], argentina2['total_deaths'])
plt.legend(['Indonesia', 'Turkey', 'Poland', 'Argentina'])        #keterangan
plt.xticks([0,30,60,90,120,150,180,210,240,270,300,330,360,390,420,450,480,510,540,570], rotation='vertical')  #posisi tanggal
plt.title("Data Covid")                                           #judul 
plt.show()
```
